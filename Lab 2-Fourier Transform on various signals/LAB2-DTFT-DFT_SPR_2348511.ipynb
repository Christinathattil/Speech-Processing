### Question 1

```
(a) Generate a basic sinusoidal signal in the time domain.
(For example, you may generate a sine wave with a frequency of 5Hz,
sampled at 1000Hz.)

(b) Plot the time-domain waveform of the signal.

(c) Compute the Discrete-Time Fourier Transform (DTFT) and plot the con-
tinuous frequency spectrum.

(d) Compute the Discrete Fourier Transform (DFT) and plot the discrete
frequency spectrum.
```

````python
# Parameters
frequency = 5  # Frequency of the sine wave in Hz
sampling_rate = 1000  # Sampling rate in Hz
duration = 1  # Duration in seconds

# Time array (sampling points)
t = np.linspace(0, duration, int(sampling_rate * duration), endpoint=False)
````

````python
# Generate the sine wave signal
amplitude = 1  # Amplitude of the sine wave
sine_wave = amplitude * np.sin(2 * np.pi * frequency * t)
````

````python
# Plot the time-domain waveform of the signal
plt.figure(figsize=(10, 4))
plt.plot(t, sine_wave)
plt.title(f'Time-Domain Waveform: {frequency} Hz Sine Wave')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.grid()
plt.xlim(0, duration)  # Limit x-axis to the duration of the signal
plt.show()
````

Define the Frequency Range

Create an array of frequency values over which to compute the DTFT. This typically ranges from -π to π (or 0 to 2π).



Screenshot 2024-09-26 at 10.36.43 PM.png

````python
# (c) Compute the DTFT
N = len(sine_wave)  # Number of samples in the signal
omega = np.linspace(-np.pi, np.pi, 1000)  # Frequency range for DTFT (continuous frequencies)
dtft = np.zeros_like(omega, dtype=complex)

for i, w in enumerate(omega):
    dtft[i] = np.sum(sine_wave * np.exp(-1j * w * np.arange(N)))

# Plot the DTFT (Continuous Frequency Spectrum)
plt.figure(figsize=(10, 4))
plt.plot(omega, np.abs(dtft))
plt.title('DTFT: Continuous Frequency Spectrum')
plt.xlabel('Frequency (radians/sample)')
plt.ylabel('Magnitude')
plt.grid()
plt.show()
````

````python
# (d) Compute the DFT
dft = np.fft.fft(sine_wave)  # Compute the DFT of the signal
dft_freqs = np.fft.fftfreq(N, 1/sampling_rate)  # Generate frequency bins

# Plot the DFT (Discrete Frequency Spectrum)
plt.figure(figsize=(10, 4))
plt.stem(dft_freqs[:N//2], np.abs(dft[:N//2]), basefmt=" ", use_line_collection=True)
plt.title('DFT: Discrete Frequency Spectrum')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.grid()
plt.show()
````

````python
#Time array

t = np.linspace(0, duration, int(sampling_rate * duration), endpoint = False)
````

````python
# a) Exponentially decaying sine wave signal

decaying_signal = amplitude * np.exp(-decay_rate * t) * np.sin( 2 * np.pi *frequency *t)
````

````python
# (b) Plot the time-domain waveform
plt.figure(figsize=(10, 4))
plt.plot(t, decaying_signal)
plt.title('Exponentially Decaying Sine Wave (Time Domain)')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.grid()
plt.show()
````

e) Analysis of the Relationship Between Time-Domain and Frequency-Domain Representations

```

In this context, we are analyzing an exponentially decaying sinusoidal signal and understanding how its time-domain waveform correlates with its frequency-domain representation (both DTFT and DFT).

#### 1. Time-Domain Waveform
The time-domain signal is given by:
\[
x(t) = A e^{-\alpha t} \sin(2\pi f t)
\]
- This is a damped sinusoid, where the oscillating sine wave is multiplied by an exponentially decaying factor \( e^{-\alpha t} \).
- As time increases, the amplitude of the sinusoid decreases exponentially.

#### Key Features of the Time-Domain Waveform:
- Oscillations: The signal oscillates with a frequency \( f \). In the example, this frequency is \( 5 \, \text{Hz} \).
- Exponential Decay: The signal's amplitude decays over time due to the factor \( e^{-\alpha t} \). The rate of decay is determined by the parameter \( \alpha \).
- Duration: Over time, the oscillations become harder to detect as the signal approaches zero.

#### 2. Frequency-Domain Representation

The frequency-domain representation describes how the signal's energy is distributed across different frequencies. In this case, we can observe the following in the DTFT (continuous spectrum) and DFT (discrete spectrum):

##### (a) Continuous Frequency Spectrum (DTFT)
The DTFT of the exponentially decaying signal shows:
- Peak around the frequency \( f \): The sinusoid oscillates with frequency \( f = 5 \, \text{Hz} \), and the DTFT shows a peak in the magnitude spectrum near this frequency.
- Broadened spectrum: Due to the exponential decay in the time domain, the spectrum is not a sharp spike but spread out over a range of frequencies. This broadening effect happens because the decaying nature of the signal introduces multiple frequency components.
- Symmetry: Since the signal is real-valued, the spectrum is symmetric around 0 frequency.

##### (b) Discrete Frequency Spectrum (DFT)
The DFT provides a sampled version of the continuous frequency spectrum. In the DFT plot:
- Discrete peaks near \( f = 5 \, \text{Hz} \): You will see clear peaks around the fundamental frequency of \( 5 \, \text{Hz} \).
- Additional frequency components: Similar to the DTFT, the spectrum is spread out across adjacent frequency bins, especially due to the decay factor in the time domain. This indicates that the signal has frequency content beyond just a single frequency \( f \).

---

### Understanding the Relationship:

1. Exponential Decay and Frequency Broadening:
   - In the time domain, the exponential decay of the signal causes the signal to be short-lived.
   - In the frequency domain, this decay introduces broadening in the spectrum. A pure sinusoidal signal (without decay) would have a narrow, sharp peak in the frequency domain at a specific frequency \( f \). However, the decaying nature of the signal spreads the energy across neighboring frequencies.
   - This is because the Fourier transform of a decaying signal has components that are distributed across a range of frequencies rather than being concentrated at one frequency.

2. Heisenberg's Uncertainty Principle:
   - This broadening of the frequency spectrum due to time-domain decay is an illustration of the Heisenberg Uncertainty Principle in signal processing: signals that are localized in the time domain (short-lived or decaying signals) tend to have broader frequency-domain representations.
   - In contrast, signals that persist longer in time (such as a steady sinusoid) have narrower frequency-domain peaks.

3. Oscillatory Nature:
   - The oscillatory nature of the time-domain signal (with frequency \( f \)) is captured in both the DTFT and DFT as a peak around the corresponding frequency.
   - The presence of multiple oscillatory components in the time-domain signal would result in multiple peaks in the frequency-domain representation.

4. Decay Rate and Frequency Spread:
   - Faster decay (higher \( \alpha \)) in the time domain leads to a wider spread in the frequency domain, because rapid decay indicates a broader range of frequencies.
   - Slower decay (lower \( \alpha \)) results in a narrower frequency peak, as the signal retains more of its oscillatory nature over time.

---

### Summary of the Relationship:
- The exponentially decaying nature of the signal in the time domain broadens its frequency-domain representation.
- The main frequency component of the signal is still clearly visible at the oscillation frequency \( f = 5 \, \text{Hz} \), but the decay introduces a spread in the frequency spectrum, leading to broader peaks.
- The DTFT provides a continuous representation, showing the smooth spread of frequency components, while the DFT shows the same phenomenon in a sampled, discrete form.

### Inference on Frequency-Domain Analysis of Various Signals Using DTFT and DFT

The frequency-domain analysis of signals provides insights into their harmonic and spectral characteristics, revealing how energy is distributed across different frequencies. Here, we infer the characteristics of different types of signals — sine wave, composite signal, exponentially decaying signal, and rectangular signal — using both the Discrete-Time Fourier Transform (DTFT) and Discrete Fourier Transform (DFT).

---

### 1. Sine Wave Signal

#### Time-Domain Characteristics:
- A sine wave is a pure periodic signal characterized by its frequency \( f \), amplitude \( A \), and phase \( \phi \).
  
#### Frequency-Domain (DTFT and DFT) Characteristics:
- DTFT: The DTFT of a sine wave shows a single sharp peak at the sine wave’s frequency \( f \) (positive and negative frequencies due to symmetry).
- DFT: The DFT provides a discrete version of the frequency spectrum, where the same peak at frequency \( f \) is observed, albeit sampled at discrete intervals determined by the number of DFT points.
  
#### Inference:
- A sine wave contains energy at a single frequency, meaning it has no harmonic content or frequency spread. It is represented as a delta function in the frequency domain, which is why both DTFT and DFT show a sharp, well-defined peak at the sine wave’s frequency.

---

### 2. Composite Signal

#### Time-Domain Characteristics:
- A composite signal is formed by adding multiple sine waves of different frequencies and amplitudes. In the time domain, it shows a combination of oscillations at different rates.

#### Frequency-Domain (DTFT and DFT) Characteristics:
- DTFT: The DTFT of a composite signal shows multiple sharp peaks, each corresponding to the individual frequencies present in the signal.
- DFT: The DFT will show discrete peaks at the individual frequencies of the composite signal, provided the signal is sampled at a sufficient rate (Nyquist criterion).
  
#### Inference:
- The frequency spectrum of a composite signal is characterized by multiple distinct peaks in both DTFT and DFT, with each peak representing a different sine wave component. The amplitude of each peak corresponds to the relative strength of the sine wave components in the composite signal.

---

### 3. Exponentially Decaying Signal

#### Time-Domain Characteristics:
- An exponentially decaying signal is a sinusoidal waveform whose amplitude decreases exponentially over time. It appears as a damped oscillation in the time domain.

#### Frequency-Domain (DTFT and DFT) Characteristics:
- DTFT: The DTFT of an exponentially decaying signal shows a broadened spectrum around the oscillation frequency, indicating the presence of multiple frequency components. The broadening happens due to the damping effect.
- DFT: The DFT also captures the spread of energy across frequencies but in a discrete form. It shows a peak at the oscillation frequency, but with additional frequencies present due to the decay.
  
#### Inference:
- Exponential decay introduces a spread of frequencies in the frequency domain. Instead of a sharp peak like in a pure sine wave, the spectrum is broader. This demonstrates that a damped signal has energy across a range of frequencies, with the oscillation frequency remaining dominant.

---

### 4. Rectangular Signal

#### Time-Domain Characteristics:
- A rectangular signal (or square wave) oscillates between two levels (e.g., \( +1 \) and \( -1 \)) with abrupt transitions, creating a non-sinusoidal periodic waveform.

#### Frequency-Domain (DTFT and DFT) Characteristics:
- DTFT: The DTFT of a rectangular signal shows a series of harmonic peaks at odd multiples of the fundamental frequency. The frequency spectrum has a slower roll-off due to the signal’s sharp edges.
- DFT: The DFT shows discrete peaks corresponding to the fundamental frequency and its odd harmonics. The amplitude of the harmonics decreases as the harmonic order increases.
  
#### Inference:
- The rectangular signal has significant harmonic content because its sharp transitions introduce high-frequency components. The frequency spectrum contains odd harmonics of the fundamental frequency. Both DTFT and DFT exhibit a peak at the fundamental frequency, with additional peaks at higher frequencies (harmonics), which are crucial in representing the discontinuous nature of the signal.

---

### Summary of Inferences:
1. Sine Wave: Has a single peak in the frequency domain at its oscillation frequency, indicating a pure tone with no harmonic content.
   
2. Composite Signal: Contains multiple distinct peaks, each corresponding to a different sine wave component, indicating that the signal is composed of several frequencies.

3. Exponentially Decaying Signal: Exhibits a broadened frequency spectrum due to the decay in the time domain. The spread shows the signal contains energy over a range of frequencies rather than just a single one.

4. Rectangular Signal: Has a rich frequency spectrum with odd harmonics of the fundamental frequency. The sharp transitions in the time domain introduce significant high-frequency content, resulting in multiple peaks in both DTFT and DFT.

---

### Conclusion:
The frequency-domain representations (DTFT and DFT) offer key insights into the behavior of different types of signals:
- Pure periodic signals (like sine waves) have simple, narrow frequency spectra.
- Signals with sharp transitions or discontinuities (like rectangular signals) have broader spectra with many harmonics.
- Composite signals combine the spectra of their individual components.
- Damped signals (like exponentially decaying signals) exhibit broad frequency content due to the time-localized nature of the signal.

By comparing DTFT and DFT, it becomes evident that DTFT provides a continuous, detailed view of the frequency content, while DFT offers a discrete approximation useful for digital processing applications.
